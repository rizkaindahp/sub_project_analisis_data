### Explore `orders_df`

```python
orders_df.sample(5)
```

```python
orders_df.describe(include="all")
```

```python
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])
orders_df['order_approved_at'] = pd.to_datetime(orders_df['order_approved_at'])
orders_df['order_delivered_carrier_date'] = pd.to_datetime(orders_df['order_delivered_carrier_date'])
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df['order_delivered_customer_date'])
orders_df['order_estimated_delivery_date'] = pd.to_datetime(orders_df['order_estimated_delivery_date'])
```

```python
# 1. Durasi antara proses pembelian hingga pengiriman
orders_df['purchase_to_delivery'] = (orders_df['order_delivered_customer_date'] - orders_df['order_purchase_timestamp']).dt.days
```

```python
# 2. Apakah ada keterlambatan dibandingkan dengan estimasi pengiriman?
orders_df['is_late'] = orders_df['order_delivered_customer_date'] > orders_df['order_estimated_delivery_date']
late_orders = orders_df[orders_df['is_late']]
total_late_orders = late_orders.shape[0]
late_percentage = (total_late_orders / orders_df.shape[0]) * 100
print(total_late_orders)  # Total keterlambatan
```

```python
# 3. Rata-rata waktu pemrosesan untuk setiap tahap
orders_df['purchase_to_approval'] = (orders_df['order_approved_at'] - orders_df['order_purchase_timestamp']).dt.seconds / 3600
orders_df['approval_to_carrier'] = (orders_df['order_delivered_carrier_date'] - orders_df['order_approved_at']).dt.days
orders_df['carrier_to_delivery'] = (orders_df['order_delivered_customer_date'] - orders_df['order_delivered_carrier_date']).dt.days
```

```python
# 4. Statistik untuk durasi waktu
processing_stats = {
    'Purchase to Approval (hours)': orders_df['purchase_to_approval'].mean(),
    'Approval to Carrier (days)': orders_df['approval_to_carrier'].mean(),
    'Carrier to Delivery (days)': orders_df['carrier_to_delivery'].mean(),
    'Purchase to Delivery (days)': orders_df['purchase_to_delivery'].mean()
}
print(processing_stats)
```

```python
# 5. Menampilkan pola dan anomali
# Contoh: Pesanan yang memakan waktu terlalu lama pada tahap tertentu
long_approval_orders = orders_df[orders_df['purchase_to_approval'] > 48]  # Approval lebih dari 48 jam
long_delivery_orders = orders_df[orders_df['carrier_to_delivery'] > 7]  # Pengiriman lebih dari 7 hari

print("\nContoh pesanan yang memakan waktu lama pada tahap Approval:")
print(long_approval_orders[['order_id', 'purchase_to_approval']].head())

print("\nContoh pesanan yang memakan waktu lama pada tahap Delivery:")
print(long_delivery_orders[['order_id', 'carrier_to_delivery']].head())
```

```python
# Menentukan kategori ukuran produk
def size_category(length):
    if length < 20:
        return 'Small'
    elif length < 50:
        return 'Medium'
    else:
        return 'Large'

products_df['size_category'] = products_df['product_length_cm'].apply(size_category)

size_group = products_df.groupby('size_category').size()
print(f"size_group : \n{size_group}")
```

```python
products_df['product_weight_g'] = pd.to_numeric(products_df['product_weight_g'], errors='coerce')

# Menentukan kategori berat
def weight_category(weight):
    if weight < 500:
        return 'Light'
    elif weight < 2000:
        return 'Medium'
    else:
        return 'Heavy'

products_df['weight_category'] = products_df['product_weight_g'].apply(weight_category)

weight_stats = products_df.groupby('weight_category')['product_weight_g'].mean()
print(weight_stats)
```

### RFM Analysis

```python
rfm_df = all_data_df.groupby(by="customer_id", as_index=False).agg({
    "order_purchase_timestamp": "max", #mengambil tanggal pesanan pembelian terakhir
    "order_id": "nunique",
    "price": "sum"
})
# Change the column names to a list with 4 elements
rfm_df.columns = ["customer_id", "max_order_purchase_timestamp", "frequency", "monetary"]
rfm_df.head()
```

```python
# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_purchase_timestamp"] = rfm_df["max_order_purchase_timestamp"].dt.date
recent_date = orders_df["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_purchase_timestamp"].apply(lambda x: (recent_date - x).days)
rfm_df.head()
```

```python
rfm_df.drop("max_order_purchase_timestamp", axis=1, inplace=True)
rfm_df.head()
```

```python
rfm_df.describe()
```

```python
rfm_df.sort_values(by="recency", ascending=True).head(5)
```

```python
rfm_df.sort_values(by="frequency", ascending=False).head(5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

sns.barplot(y="recency", x="customer_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, hue="customer_id", dodge=False,ax=ax[0]
)
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis='x', labelsize=15)

sns.barplot(y="frequency", x="customer_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, hue="customer_id", dodge=False, ax=ax[1])
# ax[1].get_legend().remove()
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)

sns.barplot(y="monetary", x="customer_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, hue="customer_id", dodge=False, ax=ax[2])
# ax[2].get_legend().remove()
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)

plt.suptitle("Best Customer Based on RFM Parameters (customer_id)", fontsize=20)
plt.show()
```

```python
rfm_df['r_rank'] = rfm_df['recency'].rank(ascending=False)
rfm_df['f_rank'] = rfm_df['frequency'].rank(ascending=True)
rfm_df['m_rank'] = rfm_df['monetary'].rank(ascending=True)

rfm_df.head()
```

```python
# normalizing the rank of the customers
rfm_df['r_rank_norm'] = (rfm_df['r_rank']/rfm_df['r_rank'].max())*100
rfm_df['f_rank_norm'] = (rfm_df['f_rank']/rfm_df['f_rank'].max())*100
rfm_df['m_rank_norm'] = (rfm_df['m_rank']/rfm_df['m_rank'].max())*100

rfm_df.drop(columns=['r_rank', 'f_rank', 'm_rank'], inplace=True)

rfm_df.head()
```

```python
rfm_df['RFM_score'] = 0.15*rfm_df['r_rank_norm']+0.28 * \
    rfm_df['f_rank_norm']+0.57*rfm_df['m_rank_norm']
rfm_df['RFM_score'] *= 0.05
rfm_df = rfm_df.round(2)
rfm_df[['customer_id', 'RFM_score']].head(7)
```

```python
rfm_df["customer_segment"] = np.where(
    rfm_df['RFM_score'] > 4.5, "Top customers", (np.where(
        rfm_df['RFM_score'] > 4, "High value customer",(np.where(
            rfm_df['RFM_score'] > 3, "Medium value customer", np.where(
                rfm_df['RFM_score'] > 1.6, 'Low value customers', 'lost customers'))))))

rfm_df[['customer_id', 'RFM_score', 'customer_segment']].head(20)
```

```python
customer_segment_df = rfm_df.groupby(by="customer_segment", as_index=False).customer_id.nunique()
customer_segment_df
```

```python
customer_segment_df['customer_segment'] = pd.Categorical(customer_segment_df['customer_segment'], [
    "lost customers", "Low value customers", "Medium value customer",
    "High value customer", "Top customers"
])
```

```python
plt.figure(figsize=(10, 5))
colors_ = ["#72BCD4", "#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(
    x="customer_id",
    y="customer_segment",
    data=customer_segment_df.sort_values(by="customer_segment", ascending=False),
    palette=colors_
)
plt.title("Number of Customer for Each Segment", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```
